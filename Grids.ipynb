```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
t=sns.load_dataset('tips')
g=sns.PairGrid(t)
g.map_diag(sns.kdeplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.kdeplot)
```
