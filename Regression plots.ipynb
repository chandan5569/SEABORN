```python
import seaborn as sns
%matplotlib inline
x=sns.load_dataset('tips')
x
```

```python
sns.lmplot(x='total_bill',y='tip',data=x)     #ROW, COLUMN AND DATA
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='Set1')
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='Set1',markers=['o','+'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='Set1',markers=['o','+'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='Set1',markers=['o','+'],col='day')
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,hue='sex',palette='Set1',markers=['o','+'],col='sex')      #Use of markers list for the different markers that is for the male is different and for the femak
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,col='time')         #use of column
```

```python
sns.lmplot(x='total_bill',y='tip',data=x,col='time',row='sex')    #Use of row and column according to the facetgrid
```
