```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
t=sns.load_dataset('tips')
sns.set_style('dark')           #set_style is used for the background effect of the plot
sns.countplot(x='sex',data=t)
```

```python
sns.set_style("darkgrid")                  #we have the darkgrid background
sns.countplot(x='sex',data=t)
```

```python
sns.set_style("ticks")
sns.countplot(x='sex',data=t)
```

```python
sns.set_style("ticks")
sns.countplot(x='sex',data=t)
sns.despine()                           #will remove the top and the right spike.
```

```python
sns.set_style("ticks")
sns.countplot(x='sex',data=t)
sns.despine(left=True,bottom=True)             #will remove the left and bottam
```

```python
#we can use the seaborn and matplotlib at the same time
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=t)
```

```python
sns.set_context('poster')         #we have the different context of styling like poster, notebook and paper etc.
sns.countplot(x='sex',data=t)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=t)
```

```python
sns.set_context('paper',font_scale=4)
sns.countplot(x='sex',data=t)
```

```python
sns.set_context('poster')
sns.countplot(x='sex',data=t)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=10,aspect=0.9,data=t)
```
