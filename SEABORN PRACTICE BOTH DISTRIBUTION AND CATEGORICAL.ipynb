```python
import seaborn as sns
%matplotlib inline
x=sns.load_dataset('tips')
x.head()
```

```python
sns.distplot(x['total_bill'])               
```

```python
sns.distplot(x['total_bill'],kde=False)     #in the Y axis we have the count 
```

```python
sns.jointplot(x='tip',y='total_bill',data=x)
```

```python
sns.jointplot(x='total_bill',y='tip',data=x)
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=x,kind='kde')
```

```python
sns.rugplot(x['total_bill'])
```

```python
sns.pairplot(data=x)
```

```python
sns.pairplot(data=x,hue='sex')
```

```python
sns.pairplot(data=x,hue='sex',palette='Set1')
```

```python
sns.kdeplot(x['total_bill'])
```

```python
sns.rugplot(x['total_bill'])
sns.kdeplot(x['total_bill'])              #combination of both kde and rug plot
```

```python
sns.barplot(x='day',y='total_bill',data=x)         #average total bill pay on each day
```

```python
import numpy as np
sns.barplot(x='day',y='total_bill',data=x,estimator=np.std) 
```

```python
sns.countplot(x='sex',data=x)
```

```python
sns.countplot(x='day',data=x)
```

```python
sns.boxplot(x='day',y='total_bill',data=x)
```

```python
sns.boxplot(x='day',y='total_bill',data=x,hue='sex')
```
