```python
import seaborn as sns
%matplotlib inline
x=sns.load_dataset('tips')
x.head()
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=x)      #Average total bill for men and female.
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=x,estimator=np.std)    #Standard value for male and female
```

```python
sns.countplot(x='sex',data=x)               #Counts the number of occurenece of male and feamle in the dataset
```

```python
sns.boxplot(x='day',y='total_bill',data=x)
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=x,palette='coolwarm')     #Use of hue feild in boxplot
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=x,palette='rainbow') 
```

```python
sns.boxplot(data=x)          #For the entire dataset here we can use the orient='h'  for horizontal
                             #here we cannot use the HUE catogorical column because her no X nad Y are present
```

```python
import seaborn as sns
%matplotlib inline                             #Revision purpose
x=sns.load_dataset('tips')
sns.pairplot(data=x)
```

```python
sns.violinplot(x='sex',y='total_bill',data=x)
```

```python
sns.boxplot(x='sex',y='total_bill',data=x)
```

```python
x.groupby('total_bill').count()
```

```python
sns.violinplot(x='day',y='total_bill',data=x)
```

```python
sns.violinplot(x='day',y='total_bill',data=x,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=x,hue='sex',split=True,palette='Set1')  
#Here split basically merges the two hue concept
```

```python
sns.stripplot(x='day',y='total_bill',data=x)           #without scattering
```

```python
sns.stripplot(x='day',y='total_bill',data=x,jitter=True)     #Use of the jitter because we want the scattering in a line
```

```python
sns.stripplot(x='day',y='total_bill',data=x,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=x,jitter=True,hue='sex',split=True,palette='Set1')
```

```python
sns.swarmplot(x='day',y="total_bill",data=x)
```

```python
sns.swarmplot(x='day',y="total_bill",data=x,hue='sex')
```

```python
sns.swarmplot(x='day',y="total_bill",data=x,hue='sex',split=True,palette='Set1')
```
